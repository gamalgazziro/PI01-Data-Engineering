```python

def get_word_count(plataforma, Keyword):
    if plataforma == 'amazon':
        return amazon[amazon['title'].str.contains(Keyword)].shape[0]
    elif plataforma == 'disney':
        return disney[disney['title'].str.contains(Keyword)].shape[0]
    elif plataforma == 'hulu':
        return hulu[hulu['title'].str.contains(Keyword)].shape[0]
    elif plataforma == 'netflix':
        return netflix[netflix['title'].str.contains(Keyword)].shape[0]
```

```python
def get_score_count(plataforma, score, anio):
    if plataforma == 'amazon':
        df = amazon[(amazon['type'] == 'movie') & (amazon['score'] > score) & (amazon['release_year'] == anio)]
        return df.shape[0]
    elif plataforma == 'disney':
        df = disney[(disney['type'] == 'movie') & (disney['score'] > score) & (disney['release_year'] == anio)]
        return df.shape[0]
    elif plataforma == 'hulu':
        df = hulu[(hulu['type'] == 'movie') & (hulu['score'] > score) & (hulu['release_year'] == anio)]
        return df.shape[0]
    elif plataforma == 'netflix':
        df = netflix[(netflix['type'] == 'movie') & (netflix['score'] > score) & (netflix['release_year'] == anio)]
        return df.shape[0]
```

```python
get_score_count('netflix', 85, 2010)
```
